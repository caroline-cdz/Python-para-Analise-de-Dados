```python
a = False
b = True
x = not a & b

# Tabela da verdade do operador & (e)
# (True & True) = True
# (True & False) = False
# (False & False) = False
# (False & True) = False
# Logo, se a é = False e b = é True, a & b = False

# Tabela da verdade do operador not (não)
# (not True) = False
# (not False) = True
# Se a & b é = False, portanto not a & b = (not False) = True

print(x)
```
